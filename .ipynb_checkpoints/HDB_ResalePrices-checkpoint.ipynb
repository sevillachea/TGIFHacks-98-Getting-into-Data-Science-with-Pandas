# Import `pandas`, `numpy` and read the CSV file into a `df`

```python
# Import libraries

# Data Cleaning
import pandas as pd
import numpy as np

# Visualisation
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

# Machine Learning
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# Read Dataset (in .csv format) into a Dataframe 

```python
# Run this in colab
# df_2012_2014 = pd.read_csv('/content/resale-flat-prices-based-on-registration-date-from-mar-2012-to-dec-2014.csv')
# df_2015_2016 = pd.read_csv('/content/resale-flat-prices-based-on-registration-date-from-jan-2015-to-dec-2016.csv')
# df_2017 = pd.read_csv('/content/resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv')

# Run this if using local device
df_2012_2014 = pd.read_csv('Data/resale-flat-prices-based-on-registration-date-from-mar-2012-to-dec-2014.csv')
df_2015_2016 = pd.read_csv('Data/resale-flat-prices-based-on-registration-date-from-jan-2015-to-dec-2016.csv')
df_2017 = pd.read_csv('Data/resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv')
```

# Concatenate Dataframes to get data from 2012 to 2017

```python
frames = [df_2012_2014, df_2015_2016, df_2017]
df = pd.concat(frames, ignore_index=True, sort=False)
display(df.head())
print(df.shape)
```

# Remove features unnecessary for your problem

```python
df_dropcols = df.drop(['block', 'street_name', 'remaining_lease', 'flat_model'], axis=1)
df_dropcols.tail()
```

# Drop NaN values from **dataframe**

```python
df_dropna = df_dropcols.dropna()
display(df_dropna.tail())
print(df_dropna.shape)
```

```python
df_dropna.info()
```

# Sorting Dataframe according to **['month', 'town', 'flat_type', 'storey_range']** (in that order)

---



```python
df_sorted = df_dropcols.sort_values(['month', 'town', 'flat_type', 'storey_range']).reset_index(drop=True)
display(df_sorted.head())
print(df_sorted.shape)
```

```python
df_sorted.info()
```

# Change dtype of features to better suit context of dataset

```python
df_sorted = df_sorted.astype({'month':'datetime64[s]', 
                              'town':'category', 
                              'flat_type':'category', 
                              'storey_range':'category',
                              'floor_area_sqm':'int64',
                              'resale_price':'int64'})
```

```python
df_sorted['lease_commence_date'] = pd.to_datetime(df_sorted['lease_commence_date'], format='%Y')
```

```python
df_sorted.info()
```

# Creating a New Feature (Feature Engineering)

```python
df_sorted['remaining_lease'] = df_sorted['month'] - df_sorted['lease_commence_date']
df_sorted.head()
```

```python
df_sorted = df_sorted.drop(['lease_commence_date'], axis=1)
```

```python
df_sorted.dtypes
```

```python
df_sorted.head()
```

```python
df_stats = df_sorted.describe()
df_stats['resale_price'] = df_stats['resale_price'].astype('int64')
df_stats
```

# Linear Regression -> Predicting "resale_price" using "floor_area_sqm"

![Image by SPSS TUTORIALS](https://spss-tutorials.com/img/simple-linear-regression-equation-linear-relation.png)

*[Image by SPSS TUTORIALS]*

# Splitting our dataset in Train and Test sets

```python
# Import LinearRegression model from Scikit-Learn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = floor_area_sqm = df_sorted['floor_area_sqm'].values.reshape(-1, 1)
y = resale_price = df_sorted['resale_price'].values.reshape(-1, 1)

# Split the Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Check the sample sizes
print("Train Set :", X_train.shape, y_train.shape)
print("Test Set  :", X_test.shape, y_test.shape)
```

# Visual Representation of the Linear Regression Model

```python
# Create a Linear Regression object
linreg = LinearRegression()

linreg.fit(X_train, y_train)
print('Intercept \t: b = ', linreg.intercept_)
print('Coefficients \t: a = ', linreg.coef_)
```

```python
# Formula for the Regression line
regline_x = X_train
regline_y = linreg.intercept_ + linreg.coef_ * X_train

# Plot the Linear Regression line
f, axes = plt.subplots(1, 1, figsize=(16, 8))
plt.scatter(X_train, y_train, edgecolors='w')
plt.plot(regline_x, regline_y, 'r-', linewidth=3)
plt.show()
```

# Prediction of Response based on the Predictor

```python
# Predict SalePrice values corresponding to GrLivArea
resale_price_pred = linreg.predict(X_test)

# Plot the Predictions
f, axes = plt.subplots(1, 1, figsize=(16, 8))
plt.scatter(X_test, y_test, color="green", edgecolors='w')
plt.scatter(X_test, resale_price_pred, color="red", edgecolors='w')
plt.show()
```
